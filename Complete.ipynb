```python
def func(frame):
    th1=cv.GaussianBlur(frame,(11,11),0)
    ret,thresh=cv.threshold(th1,200,255,cv.THRESH_BINARY)
    return thresh
```

```python
valid=[]
ans=[]
tot=[]
l=[[] for i in range(0,7)]
#result=0 reflection
#result=1 car
font = cv.FONT_HERSHEY_SIMPLEX 
def contour(d):
    count, hierarchy=cv.findContours(d.copy(),cv.RETR_TREE,cv.CHAIN_APPROX_NONE)
    c=count[:]
    tot.append(c)
    for i in count:
        x,y,w,h=cv.boundingRect(i)
        if(w>30 and h>30 and cv.contourArea(i)>1200):
            cv.rectangle(d,(x,y),(x+w,y+h),(255,0,0),10)
            text=distance(w)
            s=str(text)
            cv.putText(frame,s,(x,y),font,1,(255,0,0),4)
            l[0].append(x)
            l[1].append(y)
            l[2].append(w)
            l[3].append(h)
            l[4].append(cv.contourArea(i))
            l[5].append(w/h)
            if(cv.contourArea(i)>20000):
                l[6].append(1)
            else:
                l[6].append(0)
    valid.append(len(count))
    return d
```
